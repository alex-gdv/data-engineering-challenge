```python
import re
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("officer_snapshot.txt", names=["temp", "First Name", "Last Name", "Honours", "Care Of", "PO BOX", "First Line Address",
                                                    "Second Line Address", "Town", "County", "Country"], sep="<", header=None)
data
```

```python
def format(t):
    pattern = r'(\d{8})(\d{12})\s+(\d{8})\s+([A-Z]{1,2}[0-9R][0-9A-Z]? [0-9][A-Z]{2})?\s*(\d{6})?\s*([A-Z\.]+)?'

    match = re.match(pattern, t)
    
    company_number = match.group(1)
    person_number = match.group(2)
    appointment_date = match.group(3)
    uk_postcode = match.group(4) if match.group(4) else np.nan
    dob = match.group(5) if match.group(5) else np.nan
    title = match.group(6) if match.group(6) else np.nan
    
    return company_number, person_number, appointment_date, uk_postcode, dob, title
    
```

```python
company_numbers = []
person_numbers = []
appointment_dates = []
uk_postcodes = []
dobs = []
titles = []
for row in data.iterrows():
    c, p, a, u, d, t = format(row[1][0])
    company_numbers.append(c)
    person_numbers.append(p)
    appointment_dates.append(a)
    uk_postcodes.append(u)
    dobs.append(d)
    titles.append(t)

data["Company Number"] = company_numbers
data["Person Number"] = person_numbers
data["Appointment Date"] = appointment_dates
data["Postcode"] = uk_postcodes
data["DOB"] = dobs
data["Title"] = titles
data = data.drop(columns=["temp"])
data
```

```python
data.to_csv("officer_snapshot.csv", index=False)
```
